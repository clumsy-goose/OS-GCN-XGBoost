### 将转录本ID转换为基因ID

```python
translate = pd.read_csv('../../dataSource/osa_IRGSP_1.txlens.txt',sep='\t')
translate
```

```python
# translate = pd.read_csv('../../dataSource/osa_IRGSP_1.txlens.txt',sep='\t')
#将df中geneID列的值按translate中tx_name和gene_id对应的值进行替换
# print(translate.head())
translate['tx_name'] = translate['tx_name'].str.upper()
translate_gene = translate.set_index('tx_name')['gene_id'].to_dict()
translate_gene
# 将df中geneID列的值按transelate_gene中tx_name的对应的gene_id进行替换
df['geneID'] = df['geneID'].map(translate_gene)
```

```python
df
```

```python
#去除geneID为NaN的行
df = df.dropna(subset=['geneID'])
df
```

```python
df.to_csv('../results/translate_expression_rpkm.csv', index=False)
```

### 与蛋白质网络相关联

```python
# 将df按df['geneID']和node的第一列内联
protein = pd.read_csv('../results/protein_drop_duplicates.csv')
df = df.merge(protein, left_on='geneID', right_on='protein1', how='inner')
# 去掉多余的列
df = df.drop(columns=['protein1', 'protein2','combined_score'])
```

```python
df.to_csv('../results/rpkm_conbine_protein_network.csv', index=False)
```

### 筛选基因

```python
df = pd.read_csv('../results/rpkm_conbine_protein_network.csv')
df
```

```python
df['zero_ratio'] = 1-df.iloc[:,1:].eq(0).mean(axis=1)
# 选出不等于0的列占比80%行
df = df.loc[df['zero_ratio'] >= 0.8]
# 打印规模
print(df.shape)
df.to_csv('../results/rpkm_conbine_protein_network_0.8.csv', index=False)
```
