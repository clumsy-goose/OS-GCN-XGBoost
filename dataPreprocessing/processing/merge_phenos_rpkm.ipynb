### 缺失值处理

```python
# 数据处理规则: 每三行为一组，如果一组数据中0值数量为1个，取非0值填补缺失值，如果0值数量为2个，取非0的值的平均值填补0值，如果缺失值数量为3个，则保留0值
def fill_zero_values(group):
    # 计算0值的数量
    zero_count = (group == 0).sum()
    # 计算非0值的数量
    non_zero_count = len(group) - zero_count
    # 如果没有0值，直接返回原组
    if non_zero_count == len(group):
        return group
    # 如果有一个0值，取非0的值填补0值
    elif zero_count == 1:
        return group.replace(0, group[group != 0].mean())
    # 如果有两个0值，取非0的值的平均值填补0值
    elif zero_count == 2:
        return group.replace(0, group[group != 0].mean())
    # 如果有三个0值，则保留0值
    else:
        return group
# 读取数据
df = pd.read_csv('../results/phenos_rpkm.csv')
# 按照每三行分组    
grouped = df.groupby(df.index // 3)
# 对每组数据应用填补0值的函数
df_filled = grouped.apply(lambda x: x.apply(fill_zero_values, axis=0))
# 重置索引  
df_filled = df_filled.reset_index(drop=True)
# 保存填补后的数据
df_filled.to_csv('../results/phenos_rpkm_filled.csv', index=False)
```
