```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../results/rpkm_T.csv')
df
```

```python
data = df.iloc[:, 1:].values.flatten()
# 计算平均值
mean = np.mean(data)
# 计算中位数
median = np.median(data)    
# 计算四分位数
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)
# 计算四分位距
iqr = q3 - q1
# 计算最大值
max_value = np.max(data)
# 计算最小值
min_value = np.min(data)
# 计算标准差
std = np.std(data)
# 计算方差
var = np.var(data)
# 计算偏度
skewness = (np.sum((data - mean) ** 3) / len(data)) / (std ** 3)
# 计算峰度
kurtosis = (np.sum((data - mean) ** 4) / len(data)) / (std ** 4) - 3
# 计算变异系数
cv = std / mean
#都打印出来
print(f"均值: {mean}")
print(f"中位数: {median}")
print(f"四分位数: {q1}, {q2}, {q3}")
print(f"四分位距: {iqr}")
print(f"最大值: {max_value}")
print(f"最小值: {min_value}")
print(f"标准差: {std}")
print(f"方差: {var}")
print(f"偏度: {skewness}")
print(f"峰度: {kurtosis}")
print(f"变异系数: {cv}")
```

```python
data = pd.DataFrame(data, columns=['values'])
data
```
